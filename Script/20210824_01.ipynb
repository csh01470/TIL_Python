```python
#02. 각 데이터셋 재 할당 후 병합, 저장하기.
#(1) 저장한 데이터셋 불러와서 동적으로 할당하기
for CULTIVAR in CULTIVAR_LIST :
    globals()[f'{CULTIVAR}_data'] = pd.read_excel(f'../data/nlp/indicator_grouping/{CULTIVAR}.xlsx')

#(2) 각 데이터셋 병합하기. 
cultivar_indicator_grouping = 고시히카리_data
for CULTIVAR in CULTIVAR_LIST[1:] :
    cultivar_indicator_grouping = pd.merge(cultivar_indicator_grouping, globals()[f'{CULTIVAR}_data'], on='value', how='outer')

#(3) '덜쫀득거리' 키워드 음수 부여해서 차감시키기. 
for i in range(len(cultivar_indicator_grouping))                  :
    for k in range(len(cultivar_indicator_grouping.columns))      :
        if cultivar_indicator_grouping.iloc[i, 0] == '덜쫀득거리' :
            if cultivar_indicator_grouping.iloc[i, k] == 1        :
                cultivar_indicator_grouping.iloc[i, 0] = '쫀득하다'
                cultivar_indicator_grouping.iloc[i, k] = -1 

#(4) column 네임 변경하기. 
cultivar_indicator_grouping.rename(columns={'value' : 'indicator'}, inplace=True)

#(5) 그룹핑하기. 
cultivar_indicator_grouping = cultivar_indicator_grouping.groupby('indicator').sum()
cultivar_indicator_grouping.reset_index(inplace=True)

#(6) 처리결과 확인하기.
cultivar_indicator_grouping
```
