```python
#2021.07.12. MON
#Team_2030
 
## Data Visualization
#00. 패키지 호출
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pylab as plt
import warnings
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

#00-1. warning message ignore
warnings.filterwarnings(action='ignore')

#00-2. 시각화 옵션 설정하기. 
%matplotlib inline
```

```python
#01. 데이터셋 전처리하기.
#(1) 데이터셋 불러오기. 
viz = pd.read_csv('../data/y_vs_yhat.csv')

#(2) date 변수 날짜 형식으로 변환하기.
viz['date'] = viz['date'].astype('str')
viz['date'] =pd.to_datetime(viz['date'])

#(3) viz 데이터셋 확인하기. 
viz.info()
```

```python
#02. 시각화하기. 
#(1) target 변수 시각화하기. 
plt.figure(figsize=(16,16))
plt.plot(viz['date'], viz['y'], label='BTC_price')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
#(2) 모델 성능 시각화하기. 
plt.figure(figsize=(16,16))
plt.plot(viz['date'], viz['y'], label='BTC_price')
plt.plot(viz['date'], viz['glmnet_pred'], label='glmnet_pred_price')
plt.plot(viz['date'], viz['xgb_pred'], label='xgb_pred_price')
plt.plot(viz['date'], viz['svmpoly_pred'], label='svmpoly_pred_price')
plt.plot(viz['date'], viz['DNN_predict'], label='DNN_pred_price')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
