```python
#2021.10.19. SUN
#Team_SmokeFree

## Map Visualization Per GU
#00. 패키지 호출하기. 
import folium
import pandas as pd
import json

#01. 데이터셋 불러오기. 
#(1) 서울특별시 구 geojson 파일 불러오기. 
SEOUL_GU = json.load(open('A:/Python_Project/Team_SmokeFree/Data/00_Raw/skorea_municipalities_geo_simple.json', encoding='utf8'))

#(2) FINAL_DATASET_GU 데이터셋 불러오기.
FINAL_DATASET_GU = pd.read_csv('../DATA/FINAL_DATASET_GU_For_MV.csv', encoding='euc-kr')

#(3) 처리 결과 확인하기. 
FINAL_DATASET_GU.head()
```

```python
#02. 구별 Choropleth(단계 구분도) 구현하여 표현하기. 
#(1) 추정흡연유동인구
VALUE = '추정흡연유동인구'

map = folium.Map(
    location=[37.561, 126.986], 
    tiles='cartodb positron', 
    zoom_start=11
)

folium.Choropleth(
    geo_data=SEOUL_GU,
    data=FINAL_DATASET_GU,
    columns=('자치구', VALUE),
    key_on='feature.properties.name',
    fill_color='PuRd'
).add_to(map)

map
```
