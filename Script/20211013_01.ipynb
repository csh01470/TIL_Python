```python
#2021.10.13. WED
#Team_SmokeFree

## Data Preprocessing, BG_DONG → HJ_DONG
#00. 패키지 호출하기.
import pandas as pd 
import numpy as np 

#01. kakao_api_key.txt 파일 부르기. 
with open("A:/Python_Project/Team_SmokeFree/key/kakao_api_key.txt",mode="r") as key_file :
    kakao_api_key = key_file.read()

#02. 법정동을 행정동으로 변환하기. 
#(1) 데이터셋 불러오기. 
bus_data = pd.read_excel('../data/서울특별시_버스_pp.xlsx')
dong_data = pd.read_excel('../data/서울특별시_행정동_법정동_코드.xlsx')

#(2) 처리 결과 확인하기. 
bus_data
```

```python
dong_data
```

```python
#03. 
#(1) 법정동 무결성 검사하기. 
bus_data['IS_GOOD_BGD'] = np.nan
for i in range(len(bus_data))                          :
    for k in range(len(dong_data))                     :
        if bus_data.iloc[i, 3] == dong_data.iloc[k, 4] : 
            bus_data.iloc[i, 9] = 'YES'

#(2) 처리 결과 확인하기. 
bus_data
```

```python
#(3) IS_GOOD_BGD 컬럼에 NO 값 있는지 파악하기. 
bus_data['IS_GOOD_BGD'].value_counts()
```

```python
#(4) IS_GOOD_BGD 컬럼 삭제하기. 
bus_data.drop(['IS_GOOD_BGD'], axis=1, inplace=True)

#(5) 처리 결과 확인하기. 
bus_data
```

```python
#(6) 법정동 값을 기반으로 행정동 및 행정동_코드 변환하기. 
for i in range(len(bus_data))                          :
    for k in range(len(dong_data))                     :
        if bus_data.iloc[i, 3] == dong_data.iloc[k, 4] : 
            bus_data.iloc[i, 2] = dong_data.iloc[k, 2]
            bus_data.iloc[i, 1] = dong_data.iloc[k, 3]

#(7) 처리 결과 확인하기. 
bus_data
```

```python
#(8) 법정동 컬럼 삭제하기.
bus_data.drop(['법정동'], axis=1, inplace=True)

#(9) 처리 결과 확인하기. 
bus_data
```

```python
#(10) 처리 결과 저장하기.
bus_data.to_excel('temp_file.xlsx', index=False)
```
