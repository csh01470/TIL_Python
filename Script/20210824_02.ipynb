```python
#2021.08.18. WED
#Team_밥믈리에
 
## Indicator Scaling
#00. 패키지 호출
import pandas as pd 
import warnings 
from sklearn.preprocessing import MinMaxScaler

#00-1. warning message ignore 
warnings.filterwarnings(action='ignore')
```

```python
#01. 데이터셋 전처리하기.
#(1) 데이터셋 불러오기.
cultivar_indicator_grouping = pd.read_excel('../data/nlp/cultivar_indicator_grouping.xlsx')

#(2) indicator 컬럼 인덱스 설정하기. 
cultivar_indicator_grouping.set_index('indicator', inplace=True)

#(3) 처리 결과 확인하기.
cultivar_indicator_grouping
```

```python
#(3) 기존 데이터셋의 전치 행렬 만들기. 
cultivar_indicator_grouping_T = cultivar_indicator_grouping.T

#(4) 처리 결과 확인하기. 
cultivar_indicator_grouping_T
```

```python
#02. 데이터셋 컬럼 별로 스케일링하기.
#(1) MinMaxScaler 객체 할당하기.
scaler = MinMaxScaler()

#(2) 스케일링 처리하기. 
cultivar_indicator_scaling = pd.DataFrame(
    scaler.fit_transform(cultivar_indicator_grouping),
    columns=cultivar_indicator_grouping.columns, 
    index=cultivar_indicator_grouping.index
)

#(3) 데이터프레임으로 변환하기. 
cultivar_indicator_scaling
```

```python
#(3) 데이터셋 저장하기. 
cultivar_indicator_scaling.T.to_excel('../data/nlp/cultivar_indicator_scaling.xlsx')
```
