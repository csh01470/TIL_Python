```python
#01. 날짜에 다른 검색 url 정의
#(1) 한달에 28일, 30일, 31일인 월 정의하기. 
date_28 = [2]
date_30 = [4, 6, 9, 11]
date_31 = [1, 3, 5, 7, 8, 10, 12]

#(2) ??? (by unknown)
def datetime(y,m1,m2)               :  
    url_result=dict() 

    for i in range(m1,m2+1)         :
        month=i
        if i in date_31:
            for j in range(1,32)    :
                day=j
                month_day=str(y)+"-"+str(i)+"-"+str(j)
                date = [str(y),str(i),str(j)]
                duration = '%2Ccd_min%3A{0}%2F{1}%2F{2}%2Ccd_max%3A{3}%2F{4}%2F{5}'.format(date[1], date[2], date[0], date[1], date[2], date[0])
                base_url = 'https://www.google.com/search?q=bitcoin&biw=1005&bih=842&source=lnt&tbs=cdr%3A1'+ duration + '&tbm=nws'
                url_result[month_day]=base_url
        elif i in date_28           :
            for j in range(1,29)    :
                day=j
                month_day=str(y)+"-"+str(i)+"-"+str(j)
                date = [str(y),str(i),str(j)]
                duration = '%2Ccd_min%3A{0}%2F{1}%2F{2}%2Ccd_max%3A{3}%2F{4}%2F{5}'.format(date[1], date[2], date[0], date[1], date[2], date[0])
                base_url = 'https://www.google.com/search?q=bitcoin&biw=1005&bih=842&source=lnt&tbs=cdr%3A1'+ duration + '&tbm=nws' 
                url_result[month_day]=base_url
        else:
            for j in range(1,31)    :
                day=j
                month_day=str(y)+"-"+str(i)+"-"+str(j)
                date = [str(y),str(i),str(j)]
                duration = '%2Ccd_min%3A{0}%2F{1}%2F{2}%2Ccd_max%3A{3}%2F{4}%2F{5}'.format(date[1], date[2], date[0], date[1], date[2], date[0])
                base_url = 'https://www.google.com/search?q=bitcoin&biw=1005&bih=842&source=lnt&tbs=cdr%3A1'+ duration + '&tbm=nws'
                url_result[month_day]=base_url
    return url_result

#(3) 연도별 시작월, 종료월 지정하기. 
d1 = datetime(2018,2,12)
d2 = datetime(2019,1,12)
d3 = datetime(2020,1,12)
d4 = datetime(2021,1,7)
d1.update(d2)
d1.update(d3)
d1.update(d4)

#(4) DATE, url_list 객체 정의하기. 
DATE = [a for a in d1.keys()]
url_list = [a for a in d1.values()]
```

```python
#(1) 드라이버 모듈 호출하고 googe으로 들어가기. 
driver = webdriver.Chrome("../Module/chromedriver.exe")
driver.get("https://www.google.com")

#(2) 크롤링하기. 
company_name_list     = []
headline_list         = []
article_date_list     = []
article_num_date_list = []
contents_list         = []
for i, url in enumerate(url_list)                                   :
    time.sleep(1)
    driver.get(url)
    article_box = driver.find_element_by_xpath('//*[@id="search"]')
    article_num = article_box.find_elements_by_css_selector('g-card')
    time.sleep(0.25)
    for k in range(1,len(article_num)+1)                            :
        try                                                         :
            company_name = driver.find_element_by_xpath(f'//*[@id="rso"]/div[{k}]/g-card/div/div/div[2]/a/div/div[2]/div[1]').text
            headline     = driver.find_element_by_xpath(f'//*[@id="rso"]/div[{k}]/g-card/div/div/div[2]/a/div/div[2]/div[2]').text
            contents     = driver.find_element_by_xpath(f'//*[@id="rso"]/div[{k}]/g-card/div/div/div[2]/a/div/div[2]/div[3]/div[1]').text
            article_date = driver.find_element_by_xpath(f'//*[@id="rso"]/div[{k}]/g-card/div/div/div[2]/a/div/div[2]/div[3]/div[2]/span/span').text
            company_name_list.append(company_name)
            headline_list.append(headline)
            contents_list.append(contents)
            article_date_list.append(article_date)
            article_num_date_list.append(i)
            print(f'COMPLETE, date_num_index= {i+1}, article_index = {k}')
        except Exception                                            :
            print(f'ERROR   , date_num_index= {i+1}, article_index = {k}')

#(4) 크롤링 완료 후 드라이버 제거하기.  
driver.close()
```

```python
#03. 데이터프레임화 하기. 
df_article = pd.DataFrame({
    'company_name'     : company_name_list,
    'headline'         : headline_list,
    'contents'         : contents_list,
    'article_date'     : article_date_list,
    'article_num_date' : article_num_date_list 
})
df_article
```

```python
#04. csv 형식으로 저장하기. 
df_article.to_excel('../data/article_crawling.xlsx', index=False)
```
