```python
#2021.10.13. WED
#Team_SmokeFree

## Data Preprocessing, X, Y -> GU, HGDONG_CODE, HGDONG
#00. 패키지 호출하기.
import warnings
import pandas as pd 
import numpy as np 
import requests

#00-1. warning message ignore 
warnings.filterwarnings(action='ignore')

#01. kakao_api_key.txt 파일 부르기. 
with open("A:/Python_Project/Team_SmokeFree/key/kakao_api_key.txt",mode="r") as key_file :
    kakao_api_key = key_file.read()

#02. 법정동을 행정동으로 변환하기. 
#(1) 데이터셋 불러오기. 
SS_data = pd.read_excel('../data/01_Feature_Data/서울특별시_담배소매업_pp.xlsx')
dong_data = pd.read_excel('../data/서울특별시_행정동_법정동_코드.xlsx')

#(2) 처리 결과 확인하기. 
SS_data
```

```python
dong_data
```

```python
#(4) 구가 없는 행 추출하기. 
SS_data_no_HGD = SS_data[SS_data['자치구']=='ERROR']

#(5) 결과 확인하기. 
SS_data_no_HGD
```

```python
#03. API 이용하여 정보 추출하기. 
#(1) 주소지를 기반으로 행정동 코드, 경도, 위도 추출하여 데이터프레임에 기입하기. 
SS_data_no_HGD_x       = SS_data_no_HGD['경도'].values
SS_data_no_HGD_y       = SS_data_no_HGD['위도'].values
SS_data_no_HGD_자치구      = []
SS_data_no_HGD_행정동_코드 = []

local_url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"
for num in range(len(SS_data_no_HGD_x))                                                             :
    try                                                                                             :
        x = SS_data_no_HGD_x[num]
        y = SS_data_no_HGD_y[num]   
        url=f'{local_url}?x={x}&y={y}&input_coord=WGS84' 
    except Exception                                                                                :
        x = SS_data_no_HGD_x[num]
        y = SS_data_no_HGD_y[num]      
        url=f'{local_url}?x={x}&y={y}&input_coord=WGS84' 
    try                                                                                             : 
        REQ_RESULT = requests.get(url, headers={"Authorization":f"KakaoAK {kakao_api_key}"}).json()
        SS_data_no_HGD_자치구.append(REQ_RESULT['documents'][1]['region_2depth_name'])
        SS_data_no_HGD_행정동_코드.append(REQ_RESULT['documents'][1]['code'])
        #print(f'DONE ! index_num = {num}')
    except Exception                                                                                :
        try                                                                                         :
            x = SS_data_no_HGD_x[num]
            y = SS_data_no_HGD_y[num]   
            url=f'{local_url}?x={x}&y={y}&input_coord=WGS84' 
            REQ_RESULT = requests.get(url, headers={"Authorization":f"KakaoAK {kakao_api_key}"}).json()
            SS_data_no_HGD_자치구.append(REQ_RESULT['documents'][1]['region_2depth_name'])
            SS_data_no_HGD_행정동_코드.append(REQ_RESULT['documents'][1]['code'])
            #print(f'DONE ! index_num = {num}')
        except Exception                                                                            :
            SS_data_no_HGD_자치구.append('ERROR')
            SS_data_no_HGD_행정동_코드.append(0)
            #print(f'ERROR ! index_num = {num}')

#(2) 처리 결과 저장하기. 
SS_data_no_HGD = pd.DataFrame({
    'INDEX'              : SS_data_no_HGD['INDEX'].values,
    '자치구'             : SS_data_no_HGD_자치구,
    '행정동_코드'        : SS_data_no_HGD_행정동_코드,
    '행정동'             : SS_data_no_HGD['행정동'].values,
    '사업장명'           : SS_data_no_HGD['사업장명'].values,
    '소재지_지번_주소'   : SS_data_no_HGD['소재지_지번_주소'].values,
    '소재지_도로명_주소' : SS_data_no_HGD['소재지_도로명_주소'].values,
    '위도'               : SS_data_no_HGD['위도'].values, 
    '경도'               : SS_data_no_HGD['경도'].values
})

#(3) 처리 결과 확인하기. 
SS_data_no_HGD
```

```python
#(4) 인덱스 기반 매칭하기.
for i in range(len(SS_data)) :
    for k in range(len(SS_data_no_HGD)) :
        if SS_data['INDEX'][i] == SS_data_no_HGD['INDEX'][k] :
            SS_data['자치구'][i] = SS_data_no_HGD['자치구'][k]
            SS_data['행정동_코드'][i] = SS_data_no_HGD['행정동_코드'][k]

#(5) 처리 결과 확인하기.    
SS_data
```

```python
#04. 행정동 코드를 기반으로 행정동 값 매칭하기. 
#(1) FOR-LOOP 사용하기.
for i in range(len(SS_data))                                       :
    for k in range(len(dong_data))                                 :
        if SS_data['행정동_코드'][i] == dong_data['행정동코드'][k] : 
            SS_data['행정동'][i] = dong_data['행정동'][k]
            break

#(2) 처리 결과 확인하기. 
SS_data
```

```python
#(3) 처리 결과 저장하기.
SS_data.to_excel('temp_file.xlsx', index=False)
```
