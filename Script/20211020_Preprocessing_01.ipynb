```python
#2021.10.20. TUE
#Team_SmokeFree

## Data Preprocessing For MCLP 
#00. 패키지 호출하기. 
import warnings
import pandas as pd 
import numpy as np 
import geopandas as gpd
from fiona.crs import from_epsg

#00-1. warning message ignore 
warnings.filterwarnings(action='ignore')

#00-2. 사전 변수 설정하기.
NAME = '음식점_외식'
GU_LIST = ['금천구', '관악구', '구로구', '마포구', '은평구', '서대문구', '용산구', '중구', '동작구', '영등포구', '관악구', '서초구']
```

```python
#01. RANDOM_POINT 데이터셋 전처리하기. 
#(1) RANDOM_POINT 데이터셋 불러오기. 
RANDOM_POINT = gpd.read_file('../data/shp/RANDOM_POINT_DIFF_INTE/RANDOM_POINT_DIFF_INTE.shp', encoding='utf-8')

#(2) 필요 컬럼만 추출하기. 
RANDOM_POINT = RANDOM_POINT[['rand_point', 'SGG_NM', 'geometry']] 

#(3) 처리 결과 확인하기. 
RANDOM_POINT
```

```python
#(4) 버퍼 지정하기. 
RANDOM_POINT_BUFFER = RANDOM_POINT.buffer(100)
```

```python
#02. COUNT DATASET 불러오기. 
#(1) 데이터셋 불러오기. 
COUNT_DATASET = pd.read_excel(f'A:/Python_Project/Team_SmokeFree/Data/01_Feature_Data/서울특별시_{NAME}_pp.xlsx')

#(2) 필요 자치구만 선택하기. 
COUNT_DATASET['INJ_GU'] = 0
for i in range(len(COUNT_DATASET)) : 
  if COUNT_DATASET['자치구'][i] in GU_LIST :
    COUNT_DATASET['INJ_GU'][i] = 1

#(3) INJ_GU == 1인 값만 추출하기. 
COUNT_DATASET = COUNT_DATASET[COUNT_DATASET['INJ_GU'] == 1]

#(4) INJ_GU 컬럼 삭제하기.
COUNT_DATASET = COUNT_DATASET.drop(['INJ_GU'], axis=1)

#(5) 인덱스 초기화하기. 
COUNT_DATASET = COUNT_DATASET.reset_index()

#(6) geometry 추가하기. 
COUNT_DATASET_point = gpd.points_from_xy(COUNT_DATASET['경도'], COUNT_DATASET['위도'])

#(7) 데이터셋 재구축하기. 
COUNT_DATASET = gpd.GeoDataFrame(COUNT_DATASET[COUNT_DATASET.columns], geometry=COUNT_DATASET_point, crs=from_epsg(4326))

#(8) 좌표계 변환하기. 
COUNT_DATASET = COUNT_DATASET.to_crs(epsg=5181)

#(9) 처리 결과 확인하기. 
COUNT_DATASET
```

```python
#(6) 새 컬럼 만들기. 
RANDOM_POINT[f'{NAME}_COUNT'] = 0

#(7) 포함 여부 파악하여 계산 후, 컬럼에 값 지정하기. 
for i in range(len(RANDOM_POINT))                                                  :
    value_count_list = []
    for k in range(len(COUNT_DATASET))                                             :
        if (RANDOM_POINT_BUFFER[i].contains(COUNT_DATASET['geometry'][k])) == True :
            value_count = 'YES'
            value_count_list.append(value_count)
    RANDOM_POINT[f'{NAME}_COUNT'][i] = len(value_count_list)
    print(f'DONE ! INDEX_NUM = {i}')
```
