```python
#2021.07.10. SAT
#Team_2030
 
## NLP
#00. 패키지 호출
import numpy as np 
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#01. 뉴스기사 자연어 처리하기. 
#(1) 데이터셋 불러오기. 
article = pd.read_excel('../data/text_data/article_crawling.xlsx')

#(2) 데이터셋 살펴보기. 
article
```

```python
#(3) 데이터셋 정보 확인하기. 
article.info()
```

```python
#(4) 데이터셋 결측치 확인하기. 
article.isna().sum()
```

```python
#02. 자연어 처리하기.  
#(1) vader 객체 지정하기. 
vader = SentimentIntensityAnalyzer()

#(2) sample로 작동 확인하기. 
sample = """
The bear market in bitcoin isn’t letting up, as the original cryptoaⓙsset has fallen to its lowest level in more than two months. What’s more, other digital assets like ripple and bitcoin cash have tumbled furtherdown more than 70% from their peaks as regulatory scrutiny of the budding sector increases around the world.
"""
vader.polarity_scores(sample)
```

```python
#(3) headline 전부 감성분석하기. 
hl_senti_score_list= []
for i in range(len(article)) :
    hl_senti_score = vader.polarity_scores(article['headline'][i])['compound'] 
    hl_senti_score_list.append(hl_senti_score)

#(4) 감성분석 점수 확인하기. 
hl_senti_score_list[:10]
```

```python
#(5) contents 전부 감성분석하기. 
cts_senti_score_list= []
for i in range(len(article)) :
    cts_senti_score = vader.polarity_scores(article['contents'][i])['compound'] 
    cts_senti_score_list.append(cts_senti_score)

#(6) 감성분석 점수 확인하기. 
cts_senti_score_list[:10]
```

```python
#(5) 데이터프레임에 결합하기. 
article['hl_senti_score'] = hl_senti_score_list
article['cts_senti_score'] = cts_senti_score_list

#(6) 결합 확인하기. 
article 
```

```python
#03. article 데이터프레임 저장하기. 
article.to_excel('../data/text_data/article_senti_score.xlsx', index=False)
```
