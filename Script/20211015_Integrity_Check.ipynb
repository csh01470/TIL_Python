```python
#2021.10.15. FRI
#Team_SmokeFree

## Dataset Integrity Check
#00. 패키지 호출하기.
import pandas as pd 
import warnings

#00-1. warning message ignore 
warnings.filterwarnings(action='ignore')

#00-1. 사전 변수 설정하기.
NAME = '금연구역'
IS_GOOD_DATASET = 0

#02. 법정동을 행정동으로 변환하기. 
#(1) 데이터셋 불러오기. 
Dataset = pd.read_excel(f'../data/01_Feature_Data/서울특별시_{NAME}_pp.xlsx')
dong_data = pd.read_excel('../data/서울특별시_행정동_법정동_코드.xlsx')

#(2) 처리 결과 확인하기. 
Dataset
```

```python
#03. 행정동 무결성 검사하기. 
#(1) 작업 시행하기. 
IGH_list = []
for i in range(len(Dataset))                                    :
    for k in range(len(dong_data))                              :
        if Dataset['행정동'][i] == dong_data['행정동'][k]       : 
            IGH = f'{i}'
            IGH_list.append(IGH)
            break

#(2) 처리 결과 확인하기. 
if len(IGH_list) == len(Dataset) :
    print('COMPLETE INTEGRITY.')
else                                                            :
    for i in range(len(Dataset))                                :
        for k in range(len(dong_data))                          :
            if Dataset['행정동_코드'][i] == dong_data['행정동코드'][k] : 
                Dataset['행정동'][i] = dong_data['행정동'][k]
                break
    
    IGH_list = []
    for i in range(len(Dataset))                                :
        for k in range(len(dong_data))                          :
            if Dataset['행정동'][i] == dong_data['행정동'][k]   : 
                IGH = f'{i}'
                IGH_list.append(IGH)
                break
    IS_GOOD_DATASET =+ 1
    if len(IGH_list) == len(Dataset)                            :
        print('IMPERFECT INTEGRITY, SO THE CONVERSION CODE WORKED.')
```

```python
#04. 자치구 무결성 검사하기. 
#(1) 작업 시행하기. 
IGG_list = []
for i in range(len(Dataset))                                    :
    for k in range(len(dong_data))                              :
        if Dataset['자치구'][i] == dong_data['시군구명'][k]     : 
            IGG = f'{i}'
            IGG_list.append(IGG)
            break

#(2) 처리 결과 확인하기. 
if len(IGG_list) == len(Dataset)                                :
    print('COMPLETE INTEGRITY.')
else                                                            :
    for i in range(len(Dataset))                                :
        for k in range(len(dong_data))                          :
            if Dataset['행정동_코드'][i] == dong_data['행정동코드'][k] : 
                Dataset['자치구'][i] = dong_data['시군구명'][k]
                break
    
    IGG_list = []
    for i in range(len(Dataset))                                :
        for k in range(len(dong_data))                          :
            if Dataset['자치구'][i] == dong_data['시군구명'][k] : 
                IGG = f'{i}'
                IGG_list.append(IGG)
                break    
    IS_GOOD_DATASET =+ 1
    if len(IGG_list) == len(Dataset)                            :
        print('IMPERFECT INTEGRITY, SO THE CONVERSION CODE WORKED.')
```

```python
#05. 위도, 경도 좌표값 결측치 있는지 확인하기. 
#(1) 위도 값 파악하기. 
if Dataset['위도'].isna().sum() == 0 : 
    print('COMPLETE INTEGRITY.')
else                                 :
    IS_GOOD_DATASET =+ 1
    print('IMPERFECT INTEGRITY, SO YOU MUST DO SOME EXTRA WORK.')
```
