```python
#2021.08.01. MON
#Team_밥믈리에
 
## Data Visualization
#00. 패키지 호출
import pandas as pd 
import numpy as np 
import warnings 
import matplotlib as mpl
import matplotlib.pylab as plt

#00-1. warning message ignore 
warnings.filterwarnings(action='ignore')

#00-2. 시각화 옵션 설정하기. 
%matplotlib inline
plt.rcParams["font.family"] = 'D2Coding'
plt.rcParams["font.size"] = 14

#00-3. 지수표기법을 정수표기법으로 변환하기.
pd.options.display.float_format = '{:.5f}'.format

#00-4. 메인 변수 정의하기.
MAIN = 'PPH'
```

```python
#01. 그룹핑 및 시각화하기.
#(1) 데이터셋 불러오기.
customer_data = pd.read_excel(f'../data/preprocessing/customer_data_{MAIN}.xlsx')

#(2) 데이터셋 확인하기.
customer_data
print('Eliminated output due to data security')
```

```python
#(3) 품종을 기반으로 그룹핑 및 시각화하기.
#①그룹핑하기.
viz01 = customer_data.groupby(['cultivar']).sum()[['ATQ','TSU','ATA']].sort_values('ATA', ascending=False)
viz01
print('Eliminated output due to data security')
```

```python
#②실질거래금액을 기반으로 파이차트 시각화하기. 
plt.figure(figsize=(8,8))
plt.pie(viz01.values[:,2], labels=viz01.index, autopct='%1.2f%%', startangle=90)  
plt.axis('equal')
plt.show()
```

```python
#③총판매단위(무게)를 기반으로 파이차트 시각화하기.
plt.figure(figsize=(8,8))
plt.pie(viz01.values[:,1], labels=viz01.index, autopct='%1.2f%%', startangle=90)  
plt.axis('equal')
plt.show()
```

```python
#(4) 품종, 단위무게 별로 그룹핑 및 시각화하기. 
#①그룹핑하기.
viz02 = customer_data.groupby(['cultivar','unit_weight(kg)']).sum()[['ATQ','TSU','ATA']].sort_values('ATA', ascending=False)
viz02
print('Eliminated output due to data security')
```

```python
#②실질거래금액을 기반으로 파이차트 시각화하기. 
plt.figure(figsize=(8,8))
plt.pie(viz02.values[:,2], labels=viz02.index, autopct='%1.2f%%', startangle=90)  
plt.axis('equal')
plt.show()
```
