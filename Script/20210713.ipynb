```python
#2021.07.13. TUE
#Team_2030

## Business Modeling
#00. 패키지 호출
import numpy as np 
import pandas as pd 
```

```python
#01. BM1
#(1) 최대 수익 모델 정의하기.
def BusinessModel(start_day,term)                                                   :   
        term_date        = pd.DataFrame({'date' : pd.date_range(start_day,periods=term)})
        data_raw         = pd.read_csv('../data/BM_dataset.csv')
        data_raw['date'] = data_raw['date'].astype('str')
        data_raw['date'] = pd.to_datetime(data_raw['date'])
        BM               = pd.merge(term_date, data_raw, left_on='date', right_on='date', how='left')

        min_pred_price       = min(BM['glmnet_pred'])
        min_price            = int(BM[BM['glmnet_pred']==min_pred_price]['y'].values[0])
        min_pred_price_index = int(BM[BM['glmnet_pred']==min_pred_price]['index_num'].values[0])
        min_pred_price_day   = str(BM[BM['glmnet_pred']==min_pred_price]['date'].values[0]).split('T')[0]
        
        max_pred_price       = max(BM['glmnet_pred'])
        max_price            = int(BM[BM['glmnet_pred']==max_pred_price]['y'].values[0])
        max_pred_price_index = int(BM[BM['glmnet_pred']==max_pred_price]['index_num'].values[0])
        max_pred_price_day   = str(BM[BM['glmnet_pred']==max_pred_price]['date'].values[0]).split('T')[0]

        expected_revenue     = max_pred_price - min_pred_price
        revenue              = max_price - min_price
        if min_pred_price_index > max_pred_price_index                              :
            print('==============================')
            print(f'최대 수익 예측 모델')
            print(f'현재 날짜(관망일) : {start_day}')
            print(f'존버 기간         : {term}일')
            print(f'시작일 이후 하락장입니다.')
            print(f'BITCOIN_TRADER_by_Team_2030')
            print('==============================')
            print(' ')
        else                                                                        :
            print('==============================')
            print(f'최대 수익 예측 모델')
            print(f'현재 날짜(관망일) : {start_day}')
            print(f'존버 기간         : {term}일')
            print(f'예상 구매 날짜    : {min_pred_price_day}')
            print(f'예상 판매 날짜    : {max_pred_price_day}')
            print(f'예상 수익         : {expected_revenue:.2f}$')
            print(f'실제 수익         : {revenue:.2f}$')
            print(f'BITCOIN_TRADER_by_Team_2030')
            print('==============================')
            print(' ')
```

```python
#(2) EX1
BusinessModel('2020-12-06',10)
```

```python
#(3) EX2
BusinessModel('2020-07-07',7)
```

```python
#02. BM2
#(1) 최대 수익 구간 모델 정의하기. 
def BusinessModel_Range(start_day,range_int) :
    for k in range(2, range_int)             :
        BusinessModel(start_day,k)
```

```python
#(2) EX1
BusinessModel_Range('2020-12-06',10)
```

```python
#(3) EX2
BusinessModel_Range('2021-07-07',7)
```

```python
#03. BM3
#(1) 최초수익(Initial Revenue)모델 정의하기. 
def Business_FindIR(start_day, range_int)                                : 
    data_raw          = pd.read_csv('../data/BM_dataset.csv')
    data_raw['date']  = data_raw['date'].astype('str')
    data_raw['date']  = pd.to_datetime(data_raw['date'])
    
    min_pp_list       = []
    min_price_list    = []
    min_pp_day_list   = []
    min_pp_idx_list   = []

    max_pp_list       = []
    max_price_list    = []
    max_pp_day_list   = []
    max_pp_idx_list   = []

    er_list           = []
    revenue_list      = []
    Jonver_day_list   = [] 

    for k in range(2, range_int)                                         :
        term_date            = pd.DataFrame({'date' : pd.date_range(start_day,periods=k)})
        BM                   = pd.merge(term_date, data_raw, left_on='date', right_on='date', how='left')

        min_pred_price       = min(BM['glmnet_pred'])
        min_price            = int(BM[BM['glmnet_pred']==min_pred_price]['y'].values[0])
        min_pred_price_index = int(BM[BM['glmnet_pred']==min_pred_price]['index_num'].values[0])
        min_pred_price_day   = str(BM[BM['glmnet_pred']==min_pred_price]['date'].values[0]).split('T')[0]
        
        max_pred_price       = max(BM['glmnet_pred'])
        max_price            = int(BM[BM['glmnet_pred']==max_pred_price]['y'].values[0])
        max_pred_price_index = int(BM[BM['glmnet_pred']==max_pred_price]['index_num'].values[0])
        max_pred_price_day   = str(BM[BM['glmnet_pred']==max_pred_price]['date'].values[0]).split('T')[0]
        
        expected_revenue     = max_pred_price - min_pred_price
        Jonver_day           = max_pred_price_index - min_pred_price_index

        if min_pred_price_index < max_pred_price_index                              :
            min_pp_list.append(min_pred_price)
            min_price_list.append(min_price)
            min_pp_day_list.append(min_pred_price_day)
            min_pp_idx_list.append(min_pred_price_index)

            max_pp_list.append(max_pred_price)
            max_price_list.append(max_price)
            max_pp_day_list.append(max_pred_price_day)
            max_pp_idx_list.append(max_pred_price_index)
            
            er_list.append(expected_revenue)
            Jonver_day_list.append(Jonver_day)
    try :
        DADA = pd.DataFrame({
            'min_pp_day'        : min_pp_day_list,
            'min_pp_idx'        : min_pp_idx_list,
            'min_predict_price' : min_pp_list,
            'min_price'         : min_price_list,
            'max_pp_day'        : max_pp_day_list,
            'max_pp_idx'        : max_pp_idx_list,
            'max_predict_price' : max_pp_list,
            'max_price'         : max_price_list,
            'expected_revenue'  : er_list,
            'Jonver_day'        : Jonver_day_list
        })
        result = DADA.iloc[0,:]
        print('==============================')
        print(f'최초 수익(IR) 모델')
        print(f'현재 날짜(관망일) : {start_day}')
        print(f"수익 대기 기간    : {result['Jonver_day']}일")
        print(f"예상 매수 날짜    : {result['min_pp_day']}")
        print(f"예상 매도 날짜    : {result['max_pp_day']}")
        print(f"예상 수익         : {result['expected_revenue']:.2f}$")
        print(f'BITCOIN_TRADER_by_Team_2030')
        print('==============================')
        print(' ')
    except Exception                                                               :
        print('==============================')
        print(f'최대 수익 예측 모델')
        print(f'현재 날짜(관망일) : {start_day}')
        print(f'존버 기간         : {range_int}일')
        print(f'시작일 이후 하락장입니다.')
        print(f'BITCOIN_TRADER_by_Team_2030')
        print('==============================')
        print(' ')
```

```python
#(2) EX1
Business_FindIR('2020-12-06',10)
```
